```python
from plotnine  import *
```

```python
# reading dataset
import pandas as pd
irisUrl="https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
df = pd.read_csv(irisUrl)
df.head()
```

```python
(ggplot(df, aes('petal.width', 'petal.length', color='variety')) +
  geom_point() +
  stat_smooth(method='lm')
)
```

```python
from plotnine.data import mpg
(ggplot(mpg)
 + aes(x='manufacturer')
 + geom_bar(size=20)
 + labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)
```
